R を使って同じ問題を扱うとどうなるか。まずは対数リンク関数を使ってみる。

```r
d <- read.delim('web_traffic-nonnan.tsv', header=FALSE)
fit <- glm(V2 ~ V1, data=d, family=poisson)
summary(fit)
```

```r
plot(d$V1, d$V2)
xx <- seq(min(d$V1), max(d$V1), length=100)
lines(xx, exp(7.036 + 0.001 * xx), lwd=2)
```

次に、線形関数（ガウシアン）を使うとどうなるか。

```r
fit <- glm(V2 ~ V1, data=d, family=gaussian)
summary(fit)
```

```r
plot(d$V1, d$V2)
lines(xx, 989.0249 + 2.5962 * xx, lwd=2)
```
